```python
import pandas as pd
from spacy.lang.pt.stop_words import STOP_WORDS
from re import sub
```

```python
df1 = pd.read_excel(r'data\pt_troubleshooting.xlsx')
df2= pd.read_excel(r'data\troubleshooting.xlsx')
display(df1)
display(df2)
```

```python
df1 = df1.drop(['pag','Unnamed: 0','action','cause'],axis=1)
df2 = df2.drop('solucoes',axis=1)
list_df = [df1,df2]
```

```python
def obter_palavras_unicas(dataframe, coluna):
    palavras_unicas = []
    # Itera sobre cada linha do DataFrame
    for index, row in dataframe.iterrows():
        # Obtém o valor da coluna desejada
        texto = str(row[coluna]).lower()
        texto = sub(r'\d+', '', texto)
        texto = sub(r"[!#$%&'()*+,-./:;<=>?@[^_`{|}~]+", ' ',texto)
        texto = sub(r'\s+', ' ',texto)
        # Separa o texto em palavras
        palavras = texto.split()
        # Adiciona as palavras únicas à lista
        for palavra in palavras:
            if len(palavra) > 2 and palavra not in palavras_unicas and palavra not in STOP_WORDS and palavra != 'nan':
                palavras_unicas.append(palavra)
    return palavras_unicas
```

```python
lista1 = []
for i in list_df:
    lista2 = []
    i.astype(str)
    for j in i.columns:
        lista3 = obter_palavras_unicas(i,j)
        lista2.extend(lista3)
    lista1.extend(lista2)

new_df = pd.DataFrame(lista1,columns=['palavras'])
```

```python
new_df.to_excel('palavras.xlsx')
```

```python
import random

def substitute_letters(word):
    letters = list(word)
    index = random.randint(0, len(letters) - 1)
    letters[index] = random.choice('abcdefghijklmnopqrstuvwxyz')
    return ''.join(letters)

def insert_letter(word):
    letters = list(word)
    index = random.randint(0, len(letters))
    letters.insert(index, random.choice('abcdefghijklmnopqrstuvwxyz'))
    return ''.join(letters)

def remove_letter(word):
    letters = list(word)
    index = random.randint(0, len(letters) - 1)
    del letters[index]
    return ''.join(letters)

def transpose_letters(word):
    letters = list(word)
    index = random.randint(0, len(letters) - 2)
    letters[index], letters[index+1] = letters[index+1], letters[index]
    return ''.join(letters)

def swap_letters(word):
    letters = list(word)
    index1 = random.randint(0, len(letters) - 1)
    index2 = random.randint(0, len(letters) - 1)
    letters[index1], letters[index2] = letters[index2], letters[index1]
    return ''.join(letters)

def add_spelling_noise(word):
    letters = list(word)
    index = random.randint(0, len(letters) - 1)
    letters[index] = random.choice('abcdefghijklmnopqrstuvwxyz')
    return ''.join(letters)
```

```python
lista_to_df = []
for i in new_df['palavras']:
    qtd = []
    while len(qtd) < 100:
        substituted_word = substitute_letters(i)
        if substituted_word not in qtd:
            qtd.append(substituted_word)
        inserted_word = insert_letter(i)
        if inserted_word not in qtd:
            qtd.append(inserted_word)
        removed_word = remove_letter(i)
        if removed_word not in qtd:
            qtd.append(removed_word)
        transposed_word = transpose_letters(i)
        if transposed_word not in qtd:
            qtd.append(transposed_word)
        swapped_word = swap_letters(i)
        if swapped_word not in qtd:
            qtd.append(swapped_word)
        noisy_word = add_spelling_noise(i)
        if noisy_word not in qtd:
            qtd.append(noisy_word)
    qtd_str = ', '.join(qtd)
    lista_temp = [i,qtd_str]
    lista_to_df.append(lista_temp)
```

```python
df_erradas = pd.DataFrame(lista_to_df,columns=['palavras','erradas'])
df_erradas.to_excel('palavras_erradas.xlsx')
```
